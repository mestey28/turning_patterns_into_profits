```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv("../turning_patterns_into_profits/Data/Transactions_all_data.csv")

data.head()
```

```python
data.drop(["Unnamed: 15", "Unnamed: 16", "Unnamed: 17", "Unnamed: 18", "Unnamed: 19", "Unnamed: 20"], axis=1, inplace=True)
```

```python
tran_ct=[]

for each in data['cust_id']:
    tran_ct.append(1)
```

```python
print(len(tran_ct))
print(len(data))
```

```python
data['transaction_count']=tran_ct
data.head()
```

```python
## create column for DOB year
import datetime

year_list=[]

for each_date in data['DOB']:
    date= datetime.datetime.strptime(str(each_date), "%m/%d/%y")
    year_list.append(date.year)
```

```python
## convert gender to binary.  1=male  0=female
binary_gender=[]

for each in data['Gender']:
    if each == 'M':
        binary_gender.append(1)
    elif each =='F':
        binary_gender.append(0)
    else:
        binary_gender.append(2)
```

```python
data['birth_year']=year_list
data['gender_1=M_0=F']=binary_gender
data.head()
```

```python
## group by customer to get city codes of equal length to grouped customer_ids

data1=data.groupby(['cust_id']).mean()

data1
```

```python
## find customers with more than 6 transactions

data2=data.groupby(['cust_id']).sum()
data2=data2.loc[data2['transaction_count']>=6]
data2=data2.reset_index()
data2
```

```python
## add a count column to use for grouping counts

count=[]

for each in data2['birth_year']:
    count.append(1)
```

```python
data2["count"]=count
data2
```

```python
test_df=pd.DataFrame({
    'customer': data2['cust_id'],
    'transaction_count': data2['transaction_count'],
    'city_code': (data2['city_code'])/(data2['transaction_count']),
    'birth_year': (data2['birth_year'])/(data2['transaction_count']),
    'Gender': (data2['gender_1=M_0=F'])/(data2['transaction_count']),
    'count': data2['count']
})
```

```python
## customers with 6 or more transactions

test_df.head(10)
```

```python
gender_list=[]

for each in test_df['Gender']:
    if each == 0:
        gender_list.append('F')
    elif each == 1:
        gender_list.append('M')
    else:
        gender_list.append('n/a')
```

```python
repeat_df=pd.DataFrame({
    'customer': data2['cust_id'],
    'transaction_count': data2['transaction_count'],
    'city_code': (data2['city_code'])/(data2['transaction_count']),
    'birth_year': (data2['birth_year'])/(data2['transaction_count']),
    'gender': gender_list,
    'count': data2['count']
})
```

```python
repeat_df.head()
```

```python
## total count of repeat customers grouped by year

df1=repeat_df.groupby(['birth_year']).sum()
df1=df1.reset_index()
df1
```

```python
## create bins for age groups

decade=[]

b70=[1970,1971,1972,1973,1974,1975,1976,1977,1978,1979]
b80=[1980,1981,1982,1983,1984,1985,1986,1987,1988, 1989]
b90=[1990,1991,1992,1993,1994,1995,1996,1997,1998,1999]

for each in df1['birth_year']:
    if each in b70:
        decade.append(70)
    elif each in b80:
        decade.append(80)
    elif each in b90:
        decade.append(90)
        
```

```python
df1['birth_decade']=decade
df1
```

```python
df1.groupby(['birth_decade']).sum()
```

```python
##  The majority of repeat customers were born in the 70's and 80's
##  The company needs to market more towards the 90's kids
```
