```python
import pandas as pd
import datetime as dt
from datetime import date
```

```python
tran_df = pd.read_csv('Transactions.csv')
cust_df = pd.read_csv('Customer.csv')
cat_df = pd.read_csv('prod_cat_info.csv')
```

```python
tran_df.head()
```

```python
tran_df.tran_date = pd.to_datetime(tran_df['tran_date'], dayfirst=True).dt.date
tran_df.head()
```

```python
tran_df['prod_cat_code'] = tran_df['prod_cat_code'].astype(str).str.pad(2, side='left', fillchar='0')
tran_df['prod_subcat_code'] = tran_df['prod_subcat_code'].astype(str).str.pad(2, side='left', fillchar='0')
```

```python
tran_df['item'] = tran_df.prod_cat_code + '-' + tran_df.prod_subcat_code
tran_df.loc[tran_df['Qty'] >= 0, 'type'] = 'S'
tran_df.loc[tran_df['Qty'] < 0, 'type'] = 'R'
```

```python
tran_df.head()
```

```python
# cust_df.DOB = pd.to_datetime(cust_df['DOB']).dt.date
cust_df.DOB = pd.to_datetime(cust_df['DOB'])
```

```python
city_dict = {0.0: 'Omaha', 1.0: 'Denver', 2.0: 'Portland', 3.0: 'Las Vegas', 4.0: 'Boston' , 5.0: 'San Fransisco', 6.0: 'NewYork', 7.0: 'New Orleans', 8.0:'Miami' , 9.0: 'Nashville', 10.0: 'Bozeman'}
state_dict = {0.0: 'Nebraska', 1.0: 'Colorado', 2.0: 'Oregon', 3.0: 'Nevada', 4.0: 'Massachucetts' , 5.0: 'California', 6.0: 'NewYork', 7.0: 'Louisiana', 8.0:'Florida' , 9.0: 'Tennessee', 10.0: 'Montana'}
zip_dict = {0.0: '68124', 1.0: '80202', 2.0: '97205', 3.0: '89107', 4.0: '02113' , 5.0: '94108', 6.0: '10007', 7.0: '70116', 8.0: '33127' , 9.0: '37076', 10.0: '59717'}
cust_df['city'] = cust_df['city_code'].map(city_dict)
cust_df['state'] = cust_df['city_code'].map(state_dict)
cust_df['zip'] = cust_df['city_code'].map(zip_dict)
```

```python
cust_df.head()
```

```python
cust_df['age'] = 2020 - cust_df.DOB.dt.year
# now = dt.dt.now()
```

```python
cust_df.head()
```

```python
cat_df['prod_cat_code'] = cat_df['prod_cat_code'].astype(str).str.pad(2, side='left', fillchar='0')
cat_df['prod_sub_cat_code'] = cat_df['prod_sub_cat_code'].astype(str).str.pad(2, side='left', fillchar='0')
cat_df.head()
```

```python
tran_df.to_csv('Transactions_clean.csv', index=False)
cust_df.to_csv('Customer_clean.csv', index=False)
cat_df.to_csv('prod_cat_info_clean.csv', index=False)
```

```python
total_df = pd.merge(tran_df, cust_df, left_on='cust_id', right_on='customer_Id')
total_df = pd.merge(total_df, cat_df, left_on = ['prod_cat_code', 'prod_subcat_code'], right_on = ['prod_cat_code', 'prod_sub_cat_code'])
```
