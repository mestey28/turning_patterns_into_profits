## Use Apriori Algorithm to determine common itemsets


```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data= pd.read_csv("../turning_patterns_into_profits/Data/Transactions_all_data.csv")

data.head()
```

```python
data.drop(["Unnamed: 15", "Unnamed: 16", "Unnamed: 17", "Unnamed: 18", "Unnamed: 19", "Unnamed: 20"], axis=1, inplace=True)
```

```python
## Create product_id column for each Category+Subcategory

data["product_id"]= data['prod_cat_code'].astype(str) + "-" + data['prod_subcat_code'].astype(str)
data.head()
```

```python
##  Convert transaction id's to strings for later grouping

data['transaction_id']= data['transaction_id'].astype(str)
```

```python
## Create dataframe that represents items purchased in each transaction

basket= data.groupby(['transaction_id', 'product_id'])['Qty'].sum().unstack().reset_index().fillna(0).set_index('transaction_id')
basket
```

```python
## change quantities to 1 or zero to represent purchase or no purchase.

def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
    
basket_sets = basket.applymap(encode_units)

pd.set_option('display.max_columns', 100)
basket_sets.head()
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.0001, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

## There are no items purchased together.  No recommendations can be made based on single transactions

##  Group  by customer id rather than transaction id.
####    we can at least see which items are purchased by the same people, even if not at the same time

```python
## Create dataframe that represents items purchased by each customer id

basket2= data.groupby(['cust_id', 'product_id'])['Qty'].sum().unstack().reset_index().fillna(0).set_index('cust_id')
basket2
```

```python
## change quantities to 1 or zero to represent purchase or no purchase.

def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
    
basket_set2 = basket2.applymap(encode_units)

basket_set2.head()
```

```python
frq_cust_items = apriori(basket_set2, min_support=0.02, use_colnames=True)
rules = association_rules(frq_cust_items, metric="lift", min_threshold=1)
rules
```
